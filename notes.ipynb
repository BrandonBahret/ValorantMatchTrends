```python
import json

with open("report_data.json", "r") as fp:
    data = json.load(fp)

report = data['meta']
```

```python
from collections import defaultdict

report['agent_tree']['Jett']['by_map']['Bind']['by_rank']['gold']['teams']


tree = report['agent_tree']
ranks = set()

counter = defaultdict(int)
for agent_name in report['agent_tree']:
    if agent_name != "Jett":
        continue
    
    agent = tree[agent_name]
    for map_name in agent['by_map']:
        
        map_data = agent['by_map'][map_name]
        for rank_name in map_data['by_rank']:

            rank = map_data['by_rank'][rank_name]
            if rank.get('teams', 0) != 0:
                ranks.add(rank_name)
                
            print(f"{map_name}: {rank_name}: {rank.get('teams', 0)}")
            counter[agent_name] += rank.get('teams', 0)
            
print()
counter
```
